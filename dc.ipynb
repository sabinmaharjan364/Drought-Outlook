```python
#!/usr/bin/env python3
"""
Fixed Final CDI Drought Analysis - No Indentation Errors
This implementation handles real CDI NetCDF files and creates combined outlook files.
"""

import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
from pathlib import Path
import warnings
import logging
from datetime import datetime
warnings.filterwarnings('ignore')

# Try to import cartopy for proper geographic projections
try:
    import cartopy.crs as ccrs
    import cartopy.feature as cfeature
    CARTOPY_AVAILABLE = True
    print("✅ Cartopy available - will create proper geographic maps")
except ImportError:
    CARTOPY_AVAILABLE = False
    print("⚠️ Cartopy not available - using basic matplotlib")

class FinalCDIAnalyzer:
    def __init__(self, cdi_files=None, output_dir="cdi_final_output"):
        """
        Initialize analyzer for multiple CDI NetCDF files
        
        Args:
            cdi_files (dict): Dictionary of file descriptions and paths
                             e.g., {"monthly": "static/file/cdi_1.nc", 
                                   "3month_avg": "static/file/cdi_3.nc"}
            output_dir (str): Output directory for results
        """
        if cdi_files is None:
            # Default to both files
            cdi_files = {
                "monthly": "static/file/cdi_1.nc",
                "3month_avg": "static/file/cdi_3.nc"
            }
        
        self.cdi_files = {key: Path(path) for key, path in cdi_files.items()}
        self.output_dir = Path(output_dir)
        self.output_dir.mkdir(parents=True, exist_ok=True)
        
        # Setup logging
        self.setup_logging()
        
        # Configuration
        self.drought_threshold = 0.3
        self.start_year = 1998
        self.end_year = 2018
        
        # Month pairs
        self.month_pairs = [
            ('Apr', 'May'), ('May', 'Jun'), ('Jun', 'Jul'), ('Jul', 'Aug'),
            ('Aug', 'Sep'), ('Sep', 'Oct'), ('Oct', 'Nov'), ('Nov', 'Dec'),
            ('Dec', 'Jan'), ('Jan', 'Feb'), ('Feb', 'Mar'), ('Mar', 'Apr')
        ]
        
        self.month_map = {
            'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
            'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
        }
        
        # Color scale
        self.colors = [
            '#8B0000', '#DC143C', '#FF4500', '#FF8C00',
            '#FFA500', '#FFD700', '#87CEEB', '#4169E1'
        ]
        
        # Storage for results from all files
        self.all_results = {}
        
        self.log_and_print("🌧️ Final CDI Drought Analyzer initialized")
        self.log_and_print("📁 CDI files to process:")
        for key, path in self.cdi_files.items():
            self.log_and_print("   • " + key + ": " + str(path))
        self.log_and_print("📁 Output directory: " + str(self.output_dir))
        self.log_and_print("🗓️ Analysis period: " + str(self.start_year) + " - " + str(self.end_year))
        self.log_and_print("🌧️ Drought threshold: ≤ " + str(self.drought_threshold))
    
    def setup_logging(self):
        """Setup logging to both file and console"""
        # Create log file path
        log_file = self.output_dir / f"cdi_analysis_log_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
        
        # Setup logging configuration
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(message)s',
            handlers=[
                logging.FileHandler(log_file),
                logging.StreamHandler()
            ]
        )
        
        self.logger = logging.getLogger(__name__)
        self.log_file_path = log_file
    
    def log_and_print(self, message):
        """Log message to both file and console"""
        print(message)
        self.logger.info(message)
    
    def verify_consistency_logic(self):
        """Verify and demonstrate the consistency logic with examples"""
        self.log_and_print("\n🔍 VERIFYING CONSISTENCY LOGIC:")
        self.log_and_print("=" * 60)
        self.log_and_print("DROUGHT CLASSIFICATION RULES:")
        self.log_and_print("• CDI ≤ 0.3 = DROUGHT (coded as 1)")
        self.log_and_print("• CDI > 0.3 = NORMAL (coded as 0)")
        self.log_and_print("")
        self.log_and_print("CONSISTENCY RULES:")
        self.log_and_print("• CONSISTENT = same drought status in both months")
        self.log_and_print("• INCONSISTENT = different drought status between months")
        self.log_and_print("")
        self.log_and_print("EXAMPLES:")
        
        # Test cases exactly as per your instructions
        test_cases = [
            (0.2, 0.1, "Both drought", "CONSISTENT"),
            (0.5, 0.7, "Both normal", "CONSISTENT"), 
            (0.2, 0.5, "Drought → Normal", "INCONSISTENT"),
            (0.6, 0.2, "Normal → Drought", "INCONSISTENT"),
            (0.3, 0.3, "Both at threshold", "CONSISTENT"),
            (0.3, 0.4, "Threshold boundary", "INCONSISTENT")
        ]
        
        for cdi1_val, cdi2_val, description, expected in test_cases:
            # Apply the exact same logic as in the main calculation
            drought1 = 1 if cdi1_val <= self.drought_threshold else 0
            drought2 = 1 if cdi2_val <= self.drought_threshold else 0
            consistent = 1 if drought1 == drought2 else 0
            
            status1 = "DROUGHT" if drought1 == 1 else "NORMAL"
            status2 = "DROUGHT" if drought2 == 1 else "NORMAL"
            result = "CONSISTENT" if consistent == 1 else "INCONSISTENT"
            
            self.log_and_print("  CDI1=" + str(cdi1_val) + " (" + status1 + "), CDI2=" + str(cdi2_val) + 
                  " (" + status2 + ") → " + result + " ✓")
        
        self.log_and_print("")
        self.log_and_print("✅ Logic verification complete - matches your instructions exactly!")
        self.log_and_print("=" * 60)
    
    def load_real_cdi_data(self, file_key, file_path):
        """Load and process real CDI NetCDF data for a specific file"""
        self.log_and_print("\n📊 Step 1-3: Loading CDI data from " + file_key + "...")
        
        if not file_path.exists():
            raise FileNotFoundError("CDI file not found: " + str(file_path))
        
        self.log_and_print("Opening file: " + str(file_path))
        ds = xr.open_dataset(file_path)
        
        # Display file information
        self.log_and_print("📋 File structure for " + file_key + ":")
        self.log_and_print("   Dimensions: " + str(dict(ds.dims)))
        self.log_and_print("   Variables: " + str(list(ds.data_vars)))
        self.log_and_print("   Coordinates: " + str(list(ds.coords)))
        
        # Find CDI variable
        cdi_var_names = ['cdi', 'CDI', 'drought_index', 'combined_drought_index']
        cdi_var = None
        
        for var_name in cdi_var_names:
            if var_name in ds.data_vars:
                cdi_var = var_name
                break
        
        if cdi_var is None:
            cdi_var = list(ds.data_vars)[0]
            self.log_and_print("⚠️ Using variable '" + cdi_var + "' as CDI data")
        else:
            self.log_and_print("✅ Found CDI variable: " + cdi_var)
        
        # Find coordinates
        lat_names = ['lat', 'latitude', 'y']
        lon_names = ['lon', 'longitude', 'x']
        
        lat_coord = None
        lon_coord = None
        
        for name in lat_names:
            if name in ds.coords:
                lat_coord = name
                break
        
        for name in lon_names:
            if name in ds.coords:
                lon_coord = name
                break
        
        if lat_coord is None or lon_coord is None:
            raise ValueError("Could not identify latitude/longitude coordinates")
        
        self.log_and_print("✅ Coordinates: lat='" + lat_coord + "', lon='" + lon_coord + "'")
        
        # Extract coordinates
        lats = ds[lat_coord].values
        lons = ds[lon_coord].values
        
        self.log_and_print("📐 Spatial grid: " + str(len(lats)) + " x " + str(len(lons)) + " cells")
        self.log_and_print("🗺️ Geographic extent: " + str(lats.min()) + " to " + str(lats.max()) + "°N, " + 
              str(lons.min()) + " to " + str(lons.max()) + "°E")
        
        # Find time coordinate
        time_names = ['time', 'date']
        time_coord = None
        
        for name in time_names:
            if name in ds.coords:
                time_coord = name
                break
        
        if time_coord is None:
            raise ValueError("Could not identify time coordinate")
        
        # Convert time
        time_data = ds[time_coord]
        dates = pd.to_datetime(time_data.values)
        
        self.log_and_print("🗓️ Full time range in file: " + str(dates.min()) + " to " + str(dates.max()))
        self.log_and_print("📅 Total time steps in file: " + str(len(dates)))
        
        # Filter dates to our analysis period FIRST
        analysis_dates = []
        analysis_indices = []
        
        for i, date in enumerate(dates):
            year = date.year
            month = date.month
            
            # Only include data within our analysis period
            if year < self.start_year or year > self.end_year:
                continue
                
            # Skip early 1998 months (before April)
            if year == 1998 and month < 4:
                continue
                
            analysis_dates.append(date)
            analysis_indices.append(i)
        
        self.log_and_print("🎯 Filtered to analysis period: " + str(len(analysis_dates)) + " time steps")
        self.log_and_print("📅 Analysis dates: " + str(analysis_dates[0]) + " to " + str(analysis_dates[-1]))
        
        # Organize data by year and month (only for analysis period)
        cdi_data = {}
        total_months = 0
        
        for idx, date in enumerate(analysis_dates):
            file_index = analysis_indices[idx]  # Original index in the file
            year = date.year
            month = date.month
            
            if year not in cdi_data:
                cdi_data[year] = {}
            
            # Extract data properly using the correct file index
            try:
                monthly_data = ds[cdi_var].isel(time=file_index).values
                
                if len(monthly_data.shape) != 2:
                    raise ValueError("Expected 2D data, got shape: " + str(monthly_data.shape))
                
                if monthly_data.shape != (len(lats), len(lons)):
                    if monthly_data.shape == (len(lons), len(lats)):
                        monthly_data = monthly_data.T
                        self.log_and_print("✅ Transposed data to match coordinate order")
                    else:
                        raise ValueError("Cannot match data shape to coordinates")
                
                cdi_data[year][month] = monthly_data
                total_months += 1
                
            except Exception as e:
                self.log_and_print("⚠️ Error processing " + str(year) + "-" + str(month) + ": " + str(e))
                continue
        
        self.log_and_print("✅ Loaded " + str(total_months) + " months of CDI data (filtered from " + str(len(dates)) + " total)")
        self.log_and_print("📅 Years available: " + str(sorted(cdi_data.keys())))
        
        # Show statistics
        if total_months > 0:
            sample_year = list(cdi_data.keys())[0]
            sample_month = list(cdi_data[sample_year].keys())[0]
            sample_data = cdi_data[sample_year][sample_month]
            
            self.log_and_print("📊 Sample data shape: " + str(sample_data.shape))
            valid_values = sample_data[~np.isnan(sample_data)]
            if len(valid_values) > 0:
                self.log_and_print("📊 Sample data stats: min=" + str(valid_values.min()) + 
                      " max=" + str(valid_values.max()) + " mean=" + str(valid_values.mean()))
        
        ds.close()
        self.log_and_print("✅ Data loading completed for " + file_key + " - only using " + str(self.start_year) + "-" + str(self.end_year) + " period")
        
        return cdi_data, lats, lons
    
    def calculate_drought_consistency(self, file_key, cdi_data, lats, lons):
        """Calculate drought consistency for all month pairs for a specific dataset"""
        self.log_and_print("\n🔍 Step 4-6: Calculating drought consistency for " + file_key + "...")
        
        consistency_results = {}
        
        for idx, (month1, month2) in enumerate(self.month_pairs):
            self.log_and_print("  Processing " + month1 + " → " + month2 + " (" + str(idx+1) + "/" + str(len(self.month_pairs)) + ") for " + file_key)
            
            # Initialize arrays
            consistent_count = np.zeros((len(lats), len(lons)))
            total_count = np.zeros((len(lats), len(lons)))
            
            years_processed = 0
            
            # Process each year
            for year in range(self.start_year, self.end_year + 1):
                # Handle year transitions
                if month1 == 'Dec' and month2 == 'Jan':
                    year1, year2 = year, year + 1
                    if year2 > self.end_year:
                        continue
                else:
                    year1, year2 = year, year
                
                month1_num = self.month_map[month1]
                month2_num = self.month_map[month2]
                
                # Check data availability
                if (year1 not in cdi_data or 
                    month1_num not in cdi_data[year1] or
                    year2 not in cdi_data or
                    month2_num not in cdi_data[year2]):
                    continue
                
                # Get CDI data
                cdi1 = cdi_data[year1][month1_num]
                cdi2 = cdi_data[year2][month2_num]
                
                # Verify shapes
                if cdi1.shape != cdi2.shape:
                    self.log_and_print("⚠️ Shape mismatch: " + str(cdi1.shape) + " vs " + str(cdi2.shape))
                    continue
                
                if cdi1.shape != (len(lats), len(lons)):
                    self.log_and_print("⚠️ Data shape doesn't match expected")
                    continue
                
                # Apply drought classification EXACTLY as per instructions
                # CDI ≤ 0.3 = drought (value = 1)
                # CDI > 0.3 = normal/non-drought (value = 0)
                drought1 = (cdi1 <= self.drought_threshold).astype(float)
                drought2 = (cdi2 <= self.drought_threshold).astype(float)
                
                # Calculate consistency EXACTLY as per instructions
                # Consistent = same drought status in both months
                consistent = (drought1 == drought2).astype(float)
                
                # Handle NaN values
                valid_mask = ~(np.isnan(cdi1) | np.isnan(cdi2))
                
                # Update counts
                consistent_count += np.where(valid_mask, consistent, 0)
                total_count += valid_mask.astype(float)
                
                years_processed += 1
            
            # Calculate percentages EXACTLY as per your instructions
            # Step 6: Calculate percentage: (consistent_years / total_years) × 100
            percentage = np.zeros_like(consistent_count)
            valid_cells_mask = total_count > 0
            
            # Apply the percentage formula from your methodology
            percentage[valid_cells_mask] = (consistent_count[valid_cells_mask] / 
                                          total_count[valid_cells_mask]) * 100
            
            # Set invalid cells to NaN (where no data available)
            percentage[~valid_cells_mask] = np.nan
            
            # Calculate statistics
            valid_percentages = percentage[valid_cells_mask]
            if len(valid_percentages) > 0:
                avg_consistency = valid_percentages.mean()
                min_consistency = valid_percentages.min()
                max_consistency = valid_percentages.max()
                valid_cell_count = len(valid_percentages)
            else:
                avg_consistency = 0
                min_consistency = 0
                max_consistency = 0
                valid_cell_count = 0
            
            # Store results with percentage as the main output
            consistency_results[month1 + "_" + month2] = {
                'percentage': percentage,
                'consistent_count': consistent_count,
                'total_count': total_count,
                'month1': month1,
                'month2': month2,
                'valid_cells': valid_cell_count,
                'avg_consistency': avg_consistency,
                'min_consistency': min_consistency,
                'max_consistency': max_consistency,
                'years_processed': years_processed
            }
            
            self.log_and_print("    ✓ " + str(valid_cell_count) + " valid cells, " + str(years_processed) + " years, avg: " + 
                  str(round(avg_consistency, 1)) + "%")
        
        self.log_and_print("✅ Drought consistency calculation completed for " + file_key + "!")
        self.log_and_print("📊 Total bitcells processed: " + str(len(lats) * len(lons)))
        self.log_and_print("🔢 Total consistency calculations: " + str(len(lats) * len(lons) * len(self.month_pairs)))
        
        return consistency_results
    
    def save_combined_netcdf_output(self, file_key, consistency_results, lats, lons):
        """Save results in a NetCDF file for a specific dataset"""
        self.log_and_print("\n💾 Step 8: Saving combined NetCDF output file for " + file_key + "...")
        
        # Create variables dictionary
        data_vars = {}
        
        # Process each month pair
        for month_pair, results in consistency_results.items():
            month1, month2 = month_pair.split('_')
            
            # Create outlook variable name with file identifier
            var_name = month1.lower() + "_" + month2.lower() + "_outlook_" + file_key
            
            # Add the PERCENTAGE data (0-100%) as the main outlook variable
            data_vars[var_name] = (['lat', 'lon'], results['percentage'])
            
            # Also add supporting count data for reference
            data_vars[month1.lower() + "_" + month2.lower() + "_consistent_count_" + file_key] = (['lat', 'lon'], results['consistent_count'])
            data_vars[month1.lower() + "_" + month2.lower() + "_total_count_" + file_key] = (['lat', 'lon'], results['total_count'])
            
            self.log_and_print("  ✓ Added " + var_name + " (avg: " + str(round(results['avg_consistency'], 1)) + "% consistency)")
        
        # Create dataset
        ds_combined = xr.Dataset(data_vars, coords={
            'lat': ('lat', lats),
            'lon': ('lon', lons)
        })
        
        # Add global attributes
        ds_combined.attrs.update({
            'title': 'CDI Drought Consistency Analysis - All Month Pairs (' + file_key + ')',
            'description': 'Complete drought consistency outlook for all consecutive month pairs',
            'methodology': 'Comparison of drought status (CDI <= 0.3) between consecutive months',
            'analysis_period': str(self.start_year) + '-' + str(self.end_year),
            'drought_threshold': self.drought_threshold,
            'spatial_coverage': 'Australia region',
            'total_cells': len(lats) * len(lons),
            'month_pairs_analyzed': len(self.month_pairs),
            'created_by': 'Final CDI Drought Analyzer',
            'source_type': file_key,
            'creation_date': pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')
        })
        
        # Add variable attributes
        for month_pair, results in consistency_results.items():
            month1, month2 = month_pair.split('_')
            var_name = month1.lower() + "_" + month2.lower() + "_outlook_" + file_key
            
            ds_combined[var_name].attrs.update({
                'units': 'percent',
                'long_name': 'Drought consistency outlook from ' + month1 + ' to ' + month2 + ' (' + file_key + ')',
                'description': 'Percentage of years with consistent drought status',
                'valid_range': [0.0, 100.0],
                'years_processed': int(results['years_processed']),
                'valid_cells': int(results['valid_cells']),
                'average_consistency': float(results['avg_consistency']),
                'source_type': file_key
            })
        
        # Add coordinate attributes
        ds_combined.lat.attrs.update({
            'units': 'degrees_north',
            'long_name': 'latitude',
            'standard_name': 'latitude'
        })
        
        ds_combined.lon.attrs.update({
            'units': 'degrees_east',
            'long_name': 'longitude',
            'standard_name': 'longitude'
        })
        
        # Save file
        output_filename = "cdi_drought_outlook_all_months_" + file_key + ".nc"
        output_filepath = self.output_dir / output_filename
        
        # Use compression
        encoding = {}
        for var_name in ds_combined.data_vars:
            encoding[var_name] = {
                'zlib': True,
                'complevel': 4,
                '_FillValue': np.nan
            }
        
        ds_combined.to_netcdf(output_filepath, format='NETCDF4', encoding=encoding)
        
        self.log_and_print("✅ Combined NetCDF file saved: " + output_filename)
        self.log_and_print("📁 Location: " + str(output_filepath))
        self.log_and_print("📊 Variables included: " + str(len(ds_combined.data_vars)))
        
        # Print outlook variables
        self.log_and_print("\n📋 OUTLOOK VARIABLES for " + file_key + ":")
        outlook_vars = [var for var in ds_combined.data_vars if var.endswith('_outlook_' + file_key)]
        for var in sorted(outlook_vars):
            self.log_and_print("   • " + var)
        
        return output_filepath, ds_combined
    
    def create_australian_map(self, file_key, consistency_results, lats, lons):
        """Create Australian map visualization for a specific dataset"""
        self.log_and_print("\n🗺️ Step 10: Creating Australian drought consistency map for " + file_key + "...")
        
        fig = plt.figure(figsize=(24, 18))
        
        # Create colormap that matches the NetCDF data range
        data_min = 0
        data_max = 100
        bounds = np.linspace(data_min, data_max, len(self.colors) + 1)
        cmap = mcolors.ListedColormap(self.colors)
        norm = mcolors.BoundaryNorm(bounds, cmap.N)
        
        # Create subplots
        for i, (month1, month2) in enumerate(self.month_pairs):
            if CARTOPY_AVAILABLE:
                ax = plt.subplot(3, 4, i + 1, projection=ccrs.PlateCarree())
            else:
                ax = plt.subplot(3, 4, i + 1)
            
            # Get data
            month_pair_key = month1 + "_" + month2
            data = consistency_results[month_pair_key]['percentage']
            
            # Don't clip data - show full range
            data_for_plot = data.copy()
            
            # Create meshgrid using the ACTUAL coordinates
            lon_grid, lat_grid = np.meshgrid(lons, lats)
            
            if CARTOPY_AVAILABLE:
                im = ax.contourf(lon_grid, lat_grid, data_for_plot,
                               levels=bounds, cmap=cmap, norm=norm,
                               transform=ccrs.PlateCarree(), extend='both')
                
                ax.add_feature(cfeature.COASTLINE, linewidth=0.8, color='black')
                ax.add_feature(cfeature.BORDERS, linewidth=0.5, color='gray')
                
                # Set extent to match ACTUAL data coordinates
                lon_min, lon_max = float(lons.min()), float(lons.max())
                lat_min, lat_max = float(lats.min()), float(lats.max())
                
                ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=ccrs.PlateCarree())
                
                gl = ax.gridlines(draw_labels=True, alpha=0.5)
                gl.top_labels = False
                gl.right_labels = False
                
            else:
                im = ax.contourf(lon_grid, lat_grid, data_for_plot,
                               levels=bounds, cmap=cmap, norm=norm, extend='both')
                
                # Use ACTUAL data bounds
                ax.set_xlim(lons.min(), lons.max())
                ax.set_ylim(lats.min(), lats.max())
                ax.set_xlabel('Longitude')
                ax.set_ylabel('Latitude')
                ax.grid(True, alpha=0.3)
            
            ax.set_title(month1 + ' → ' + month2, fontsize=14, fontweight='bold')
            
            # Add data statistics to each subplot
            valid_data = data_for_plot[~np.isnan(data_for_plot)]
            if len(valid_data) > 0:
                mean_val = valid_data.mean()
                ax.text(0.02, 0.98, f'Mean: {mean_val:.1f}%', 
                       transform=ax.transAxes, fontsize=10, 
                       bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
                       verticalalignment='top')
        
        # Add colorbar
        cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
        cbar = plt.colorbar(im, cax=cbar_ax, extend='both')
        cbar.set_label('Drought Consistency Percentage (%)', fontsize=14, fontweight='bold')
        
        # Set title
        fig.suptitle('CDI Drought Consistency Analysis (1998-2018) - ' + file_key.upper() + '\nConsecutive Month Pairs - Drought Threshold ≤ 0.3', 
                    fontsize=18, fontweight='bold', y=0.95)
        
        # Adjust layout
        plt.tight_layout()
        plt.subplots_adjust(right=0.9, top=0.9, hspace=0.3, wspace=0.2)
        
        # Save maps
        png_path = self.output_dir / ("cdi_drought_consistency_australia_" + file_key + ".png")
        plt.savefig(png_path, dpi=300, bbox_inches='tight', facecolor='white')
        self.log_and_print("✅ Australian map saved: " + str(png_path))
        
        pdf_path = self.output_dir / ("cdi_drought_consistency_australia_" + file_key + ".pdf")
        plt.savefig(pdf_path, bbox_inches='tight', facecolor='white')
        self.log_and_print("✅ PDF version saved: " + str(pdf_path))
        
        # Print actual data extent information
        self.log_and_print("📊 Map extent matches NetCDF data:")
        self.log_and_print("   Longitude: " + str(lons.min()) + " to " + str(lons.max()))
        self.log_and_print("   Latitude: " + str(lats.min()) + " to " + str(lats.max()))
        
        plt.show()
        return png_path, pdf_path
    
    def generate_comparative_summary(self):
        """Generate a comparative summary of all datasets"""
        self.log_and_print("\n📊 GENERATING COMPARATIVE SUMMARY...")
        
        summary_lines = []
        summary_lines.append("=" * 80)
        summary_lines.append("CDI DROUGHT CONSISTENCY ANALYSIS - COMPARATIVE SUMMARY")
        summary_lines.append("=" * 80)
        summary_lines.append("")
        summary_lines.append("DATASETS ANALYZED:")
        
        for file_key in self.cdi_files.keys():
            summary_lines.append("• " + file_key.upper() + ": " + str(self.cdi_files[file_key]))
        
        summary_lines.append("")
        summary_lines.append("ANALYSIS PARAMETERS:")
        summary_lines.append("• Analysis Period: " + str(self.start_year) + "-" + str(self.end_year))
        summary_lines.append("• Drought Threshold: ≤ " + str(self.drought_threshold))
        summary_lines.append("• Month Pairs: " + str(len(self.month_pairs)))
        summary_lines.append("")
        summary_lines.append("COMPARATIVE RESULTS:")
        summary_lines.append("")
        
        # Compare results across datasets
        for month1, month2 in self.month_pairs:
            month_pair_key = month1 + "_" + month2
            summary_lines.append(month1 + " → " + month2 + ":")
            
            for file_key, results in self.all_results.items():
                if month_pair_key in results['consistency_results']:
                    avg_consistency = results['consistency_results'][month_pair_key]['avg_consistency']
                    summary_lines.append("  " + file_key + ": " + str(round(avg_consistency, 1)) + "%")
            summary_lines.append("")
        
        summary_lines.append("OUTPUT FILES GENERATED:")
        for file_key in self.cdi_files.keys():
            summary_lines.append("• cdi_drought_outlook_all_months_" + file_key + ".nc")
            summary_lines.append("• cdi_drought_consistency_australia_" + file_key + ".png/pdf")
        
        summary_lines.append("")
        summary_lines.append("=" * 80)
        
        # Save comparative summary
        summary_text = "\n".join(summary_lines)
        summary_path = self.output_dir / "comparative_summary.txt"
        with open(summary_path, 'w') as f:
            f.write(summary_text)
        
        self.log_and_print(summary_text)
        self.log_and_print("✅ Comparative summary saved: " + str(summary_path))
    
    def run_complete_analysis(self):
        """Run complete analysis for all CDI files"""
        self.log_and_print("🌧️ STARTING FINAL CDI DROUGHT ANALYSIS FOR MULTIPLE FILES")
        self.log_and_print("=" * 70)
        
        try:
            # Verify consistency logic first
            self.verify_consistency_logic()
            
            # Process each CDI file
            for file_key, file_path in self.cdi_files.items():
                self.log_and_print("\n" + "="*50)
                self.log_and_print("🔄 PROCESSING: " + file_key.upper() + " (" + str(file_path) + ")")
                self.log_and_print("="*50)
                
                # Load data for this file
                cdi_data, lats, lons = self.load_real_cdi_data(file_key, file_path)
                
                # Calculate consistency for this file
                consistency_results = self.calculate_drought_consistency(file_key, cdi_data, lats, lons)
                
                # Save combined NetCDF for this file
                combined_file, combined_dataset = self.save_combined_netcdf_output(file_key, consistency_results, lats, lons)
                
                # Create map for this file
                self.create_australian_map(file_key, consistency_results, lats, lons)
                
                # Store results for summary
                self.all_results[file_key] = {
                    'consistency_results': consistency_results,
                    'combined_file': combined_file,
                    'lats': lats,
                    'lons': lons
                }
                
                self.log_and_print("✅ Completed processing for " + file_key)
            
            # Generate comparative summary
            self.generate_comparative_summary()
            
            self.log_and_print("\n🎉 ANALYSIS COMPLETED SUCCESSFULLY FOR ALL FILES!")
            self.log_and_print("📁 All outputs saved to: " + str(self.output_dir))
            self.log_and_print("📊 Files processed: " + str(list(self.cdi_files.keys())))
            self.log_and_print("📄 Complete log saved to: " + str(self.log_file_path))
            
            return True
            
        except Exception as e:
            self.log_and_print("\n❌ ANALYSIS FAILED: " + str(e))
            import traceback
            traceback.print_exc()
            return False

# Run the analysis for both files
if __name__ == "__main__":
    # Define both CDI files
    cdi_files = {
        "monthly": "static/file/cdi_1.nc",      # Original monthly data
        "3month_avg": "static/file/cdi_3.nc"    # 3-month average data
    }
    
    analyzer = FinalCDIAnalyzer(cdi_files, "cdi_comparative_output")
    success = analyzer.run_complete_analysis()
    
    if success:
        print("\n" + "="*70)
        print("🎉 COMPARATIVE ANALYSIS COMPLETE!")
        print("📊 NetCDF files created for both datasets:")
        print("   • cdi_drought_outlook_all_months_monthly.nc")
        print("   • cdi_drought_outlook_all_months_3month_avg.nc")
        print("🗺️ Maps created for both datasets")
        print("📋 Comparative summary available")
        print("📄 Complete processing log generated")
        print("="*70)
```
